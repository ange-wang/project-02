```python
#Clean Species Column 
cleaner_shark_list = []
for shark in clean_shark_list:
    clean_shark = shark.strip()
    clean_shark = clean_shark.rstrip()
    clean_shark = clean_shark.replace('"','')
    clean_shark = clean_shark.replace('?','')
    clean_shark = clean_shark.replace("'","")
    clean_shark = clean_shark.replace("”","")
    clean_shark = clean_shark.replace("“","")
    clean_shark = clean_shark.lower()
    clean_shark = clean_shark.replace("sharks","shark")
    clean_shark = clean_shark.replace("shark ","shark")
    clean_shark = clean_shark.replace("pup","")
    clean_shark = clean_shark.replace("a ","")
    clean_shark = clean_shark.replace("unknown","unidentified")
    clean_shark = clean_shark.replace("shark ","shark")
    if (   
           clean_shark.find("small") != -1 or clean_shark.find("large") != -1 or 
           clean_shark.find("juvenile") != -1 or clean_shark.find("little") != -1 or
           clean_shark.find("unidentified") != -1 or clean_shark.find("miami") != -1 or
           clean_shark.find("most") != -1 or clean_shark.find("young") != -1
        ):
        clean_shark = "unidentified"
    if not clean_shark:
        clean_shark = "unidentified"
    if clean_shark == "shark":
        clean_shark = "unidentified"
    if clean_shark.find("sand") != -1:
        clean_shark = "sand tiger shark"
    if clean_shark.find("hammer") != -1:
        clean_shark = "hammerhead shark"
    if clean_shark.find("wobbegong") != -1:
        clean_shark = "wobbegong shark"
    if clean_shark.find("grey") != -1 or clean_shark.find("gray") != -1:
        clean_shark = "grey reef shark"
    if clean_shark == "bonitsharkk":
        clean_shark = "mako shark"
    if clean_shark.find("dog") != -1:
        clean_shark = "spurdog"
    if clean_shark == "horn shar": 
        clean_shark = "horn shark"
        
    cleaner_shark_list.append(clean_shark)
new_sharky["new_sp"] = cleaner_shark_list
new_sharky.drop(new_sharky.columns[14],axis=1,inplace=True)
new_sharky.rename(columns={"new_sp":"Species"}, inplace = True)
new_sharky.to_csv("clean_shark_dataV2.csv", encoding="utf-8", index=False)
```

Scrape the PDF files in the designated folders

```python
moon_phase_list =[]
moon_percent_list = []
case_id_list = []
#define Directory where pdf files are located
directory ="SharkAttacks"
print ("Starting Scrape")
for filename in os.listdir(directory):
    phase = False
    case_found = False
    #look for pdf file in difined folder
    if filename.endswith(".pdf"):
        #print (filename)
        splitcase = []
        if "ND-" in filename:
            splitcase = filename.split("-")
            sharkcase = splitcase[0] + "-" + splitcase[1]
            case_id_list.append(sharkcase)
            case_found = True
            #print (sharkcase)
        else:
            splitcase = filename.split("-")
            sharkcase = splitcase[0] 
            case_id_list.append(sharkcase)
            case_found = True
            #print (sharkcase)
        
        pdffile = os.path.join(directory, filename)
        #get file
        with open(pdffile,'rb') as f:
            text = extract_text(f)
            text_list = []
            text_list = text.splitlines()
            
            #loop through each text in list to find keywords
            for moon in text_list:
                #look for moonphase 
                if "MOON PHASE" in moon:
                    try:
                        #get moonphase
                        splitmoon = []
                        splitmoon = moon.split(",")
                        moonphase_split = splitmoon[0].split(": ")
                        #get moon visibility percent
                        #monnum = splitmoon[1].split(" ")
                        #moonpercent = monnum[1]
                        #get moon phase name
                        moonphase = moonphase_split[1]
                        moon_phase_list.append(moonphase)
                        #moon_percent_list.append(moonpercent)
                        phase = True
                        break
                    except:
                        phase = False
                        
                else:
                    continue
                    
    else:
        continue
    #if file have no moonphase append empty string
    if case_found == False:
        sharkcase = ""
        case_id_list.append(sharkcase)
    if phase == False:
        moonphase = ""
        moon_phase_list.append(moonphase)
    text = ""
    
print("Scrape Done!")
```

```python
print(len(case_id_list))
print(len(moon_phase_list))
pfd_scrape = pd.DataFrame({
    "Case Number":case_id_list,
    "Moon Phase":moon_phase_list
})
moon_phase = pfd_scrape["Moon Phase"].value_counts()
pfd_scrape.to_csv("mooncycle_pdfscraped.csv", encoding="utf-8", index=False)
moon_phase.head(60)
```

```python
moon_phase_df = pd.read_csv("mooncycle_pdfscraped.csv")
moon_phase_df = moon_phase_df.fillna("Undefined")
moon_phase  = moon_phase_df.iloc[:, 1]
moon_phase_list = []
for moon in moon_phase:
    clean_moon = moon
    #clean_moon = moon.strip()
    #clean_moon = moon.rstrip()
    if clean_moon.find("Third") != -1:
        clean_moon = "Third Quarter"
        moon_phase_list.append(clean_moon)
        continue
    elif clean_moon.find("Waning") != -1:
        clean_moon = "Waning Gibbous"
        moon_phase_list.append(clean_moon)
        continue
    elif clean_moon.find("Full") != -1:
        clean_moon = "Full Moon"
        moon_phase_list.append(clean_moon)
        continue
    elif clean_moon.find("Waxing") != -1:
        clean_moon = "Waxing Gibbous"
        moon_phase_list.append(clean_moon)
        continue
    elif clean_moon.find("First") != -1:
        clean_moon = "First Quarter"
        moon_phase_list.append(clean_moon)
        continue
    elif clean_moon.find("Waxing") != -1:
        clean_moon = "Waxing Crescent"
        moon_phase_list.append(clean_moon)
        continue
    elif clean_moon.find("New") != -1:
        clean_moon = "New Moon"
        moon_phase_list.append(clean_moon)
        continue
    elif clean_moon.find("Waning C") != -1 or clean_moon.find("Last") != -1:
        clean_moon = "Waning Crescent"
        moon_phase_list.append(clean_moon)
        continue
    else:
        clean_moon = "Undefined"
        moon_phase_list.append(clean_moon)
#moon_phase_list.append(clean_moon)
print(len(moon_phase_list))
moon_phase_df["Moon Phase"] = moon_phase_list
moon_phase_df["Moon Phase"].value_counts()
```
